```python
h_RecDiMuonM_Nocut_1 = ROOT.TH1F()
h_RecDiMuonM_pt_1 = ROOT.TH1F()
h_RecDiMuonM_eta_1 = ROOT.TH1F()
h_RecDiMuonM_iso_1 = ROOT.TH1F()
h_RecDiMuonM_dxy_1 = ROOT.TH1F()
h_RecDiMuonM_1 = ROOT.TH1F()
h_RecDiMuonM_Global_1 = ROOT.TH1F()

h_RecDiMuonM_Nocut_1 = f1.Get("demo/h_RecDiMuonM_Nocut")
h_RecDiMuonM_pt_1 = f1.Get("demo/h_RecDiMuonM_pt")
h_RecDiMuonM_eta_1 = f1.Get("demo/h_RecDiMuonM_eta")
h_RecDiMuonM_iso_1 = f1.Get("demo/h_RecDiMuonM_iso")
h_RecDiMuonM_dxy_1 = f1.Get("demo/h_RecDiMuonM_dxy")
h_RecDiMuonM_1 = f1.Get("demo/h_RecDiMuonM")
h_RecDiMuonM_Global_1 = f1.Get("demo/h_RecDiMuonM_Global")

# set the scale factor
lumi1 = input("input the int. lumi. of dataset 1: ")

lumi1 = float(lumi1)

scale1 = lumi_standard/lumi1

# scale histograms
for i in [h_RecDiMuonM_Nocut_1,h_RecDiMuonM_pt_1,h_RecDiMuonM_eta_1,h_RecDiMuonM_iso_1,h_RecDiMuonM_dxy_1,h_RecDiMuonM_1,h_RecDiMuonM_Global_1]:
    i.Scale(scale1)
    entries = i.GetEntries()
    i.SetEntries(entries * scale1)
    hist_uplim = i.GetNbinsX() + 1
    for a in range(1,hist_uplim):
        i.SetBinError(a,0)

h_RecDiMuonM_Nocut = h_RecDiMuonM_Nocut_1
h_RecDiMuonM_pt = h_RecDiMuonM_pt_1
h_RecDiMuonM_eta = h_RecDiMuonM_eta_1
h_RecDiMuonM_iso = h_RecDiMuonM_iso_1
h_RecDiMuonM_dxy = h_RecDiMuonM_dxy_1
h_RecDiMuonM = h_RecDiMuonM_1
h_RecDiMuonM_Global = h_RecDiMuonM_Global_1
```
